##cual es la palabra mas corta con el valor 62?

```python
##primero se hace el filtrado con loc y luego se ordena de menor a mayor segun el char count
df.loc[df["value"]==62].sort_values("char count")
```

##crear una columna ratio que seria la division entre el valor y el largo

```python
df["ratio"]=df["value"]/df["char count"]
df
```

##cual es el ratio maximo

```python
df["ratio"].max()
```

## que palabra tienen n el maximo ratio

```python
df.loc[df["ratio"]==10]
```

## el LOC puede ser reemplazado por query

```python
df.query ("ratio==10")
```

##si queremos saber cuantas palabras tienen el ratio =10 se hace un shape del query

```python
df.query ("ratio==10").shape
```

##de todas las palabra con ratio = 10 cual tiene el valor mas alto? entonces al query se le hace sort

```python
df.query ("ratio==10").sort_values(by="value", ascending = False)
```

##de las palabra con un value de 35 cual es la que tiene menos char count

```python
df.query ("value==45").sort_values(by="char count", ascending = True)
```

##mostrar las palabra que tienen un char count > que la media

```python
media= df["char count"].mean()
print (media)
```

```python
#si queremos hacer esto con query tenemos que tener en cuenta que no toma espacios en el
#nombre de la columna y la variable se referencia con @
df.query ("`char count`> @media")
```

```python
df.loc[df["char count"]>media]   #loc no tiene dificultades con el espacio y las variables
```
